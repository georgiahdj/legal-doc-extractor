```python
from ollama_extractor import safe_extract_from_page
from extractor import pdf_to_images, merge_results, save_results
from evaluator import count_extracted_fields, check_ground_truth_accuracy
from prompts import PROMPTS
import json, time

images = pdf_to_images('data/simbolaio-agorapolisias-public.pdf')
print(f"Συνολικές σελίδες: {len(images)}")

all_results = []
all_times = []
total_start = time.time()

for i, img in enumerate(images):
    print(f"Σελίδα {i+1}/{len(images)}...", end=" ")
    start = time.time()
    result = safe_extract_from_page(img, PROMPTS['chain_of_thought'])
    elapsed = time.time() - start
    all_times.append(elapsed)
    all_results.append(result)
    fields = count_extracted_fields(result)
    print(f"{elapsed:.1f}s | fields: {fields['total_filled']}")

final = merge_results(all_results)
total_time = time.time() - total_start

# Metrics
fields = count_extracted_fields(final)
gt = check_ground_truth_accuracy(final)

print(f"\n=== ΜΕΤΡΙΚΕΣ ===")
print(f"Συνολικός χρόνος: {total_time:.1f}s")
print(f"Μέσος χρόνος/σελίδα: {sum(all_times)/len(all_times):.1f}s")
print(f"Fields extracted: {fields['total_filled']}")
print(f"GT Score: {gt['score']}/5")
print(f"GT Details: {gt}")

print(f"\n=== ΤΕΛΙΚΟ JSON ===")
print(json.dumps(final, ensure_ascii=False, indent=2))

save_results(final, 'final_results_full.json')
print("\nΑποθηκεύτηκε στο final_results_full.json")
```
